```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Add, Lambda, LeakyReLU, Flatten, Dense
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input
from tensorflow.python.keras.layers import PReLU
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy, MeanAbsoluteError
from tensorflow.keras.optimizers.schedules import PiecewiseConstantDecay
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Mean
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from sklearn.model_selection import train_test_split
from PIL import Image
import time

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
print(tf.config.list_physical_devices('GPU'))
```

```python
lr_images_path = '../scratch/lung_images/data/lr_images'
hr_images_path = '../scratch/lung_images/data/hr_images'
```

```python
lr_filenames = sorted(os.listdir(lr_images_path))[:100]
hr_filenames = sorted(os.listdir(hr_images_path))[:100]

# Verify that the filenames match
if lr_filenames != hr_filenames:
    raise ValueError("Filenames in lr_images and hr_images do not match!")
```

```python
lr_images = []
hr_images = []


for img_name in lr_filenames:
    lr_img = Image.open(os.path.join(lr_images_path, img_name))
    hr_img = Image.open(os.path.join(hr_images_path, img_name))
    
    # Convert grayscale images to RGB format
    if lr_img.mode != 'RGB':
        lr_img = lr_img.convert('RGB')
        
    if hr_img.mode != 'RGB':
        hr_img = hr_img.convert('RGB')
    
    lr_img = np.asarray(lr_img)
    hr_img = np.asarray(hr_img)
    
    lr_images.append(lr_img)
    hr_images.append(hr_img)
```

```python
lr_images_array = np.stack(lr_images)
hr_images_array = np.stack(hr_images)

print("LR Images Array Shape:", lr_images_array.shape)
print("HR Images Array Shape:", hr_images_array.shape)
```

```python
# Split the data into training, validation, and test sets
random_seed = 42 

train_lr_images, val_test_lr_images, train_hr_images, val_test_hr_images = train_test_split(
    lr_images_array, hr_images_array, test_size=0.3, random_state=random_seed
)

val_lr_images, test_lr_images, val_hr_images, test_hr_images = train_test_split(
    val_test_lr_images, val_test_hr_images, test_size=0.5, random_state=random_seed
)
```

```python
train_data_gen_args = dict(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

train_lr_image_datagen = ImageDataGenerator(**train_data_gen_args)
train_hr_image_datagen = ImageDataGenerator(**train_data_gen_args)

# Fit the generators only on the training data (with augmentation)
train_lr_image_datagen.fit(train_lr_images, augment=True, seed=random_seed)
train_hr_image_datagen.fit(train_hr_images, augment=True, seed=random_seed)

# Create generator instances
batch_size = 16

train_lr_generator = train_lr_image_datagen.flow(train_lr_images, batch_size=batch_size)
train_hr_generator = train_hr_image_datagen.flow(train_hr_images, batch_size=batch_size)
```

```python
num_val_samples = val_lr_images.shape[0]

# Calculate the number of batches
num_val_batches = num_val_samples // batch_size

# Reshape the data into batches
val_lr_batches = val_lr_images[:num_val_batches * batch_size].reshape(
    num_val_batches, batch_size, *val_lr_images.shape[1:])

val_hr_batches = val_hr_images[:num_val_batches * batch_size].reshape(
    num_val_batches, batch_size, *val_hr_images.shape[1:])
```

```python
def normalize(x):
    return x / 255.0

def denormalize(x):
    return (x + 1) * 127.5

def pixel_shuffle(scale):
    return lambda x : tf.nn.depth_to_space(x, scale)

def upsample(x_in, num_filters):
    x = Conv2D(num_filters, kernel_size = 3, padding = 'same')(x_in)
    x = Lambda(pixel_shuffle(2))(x)
    return PReLU(shared_axes=[1,2])(x)

def residual_block(block_input, num_filters, momentum=0.8):
    x = Conv2D(num_filters, kernel_size=3, padding='same')(block_input)
    x = BatchNormalization(momentum=momentum)(x)
    x = PReLU(shared_axes=[1, 2])(x)
    x = Conv2D(num_filters, kernel_size=3, padding='same')(x)
    x = BatchNormalization(momentum=momentum)(x)
    x = Add()([block_input, x])
    return x

def build_srresnet(num_filters=64, num_res_blocks=16):
    
    num_upsamples = 2

    lr = Input(shape=(64, 64, 3))
    x = Lambda(normalize)(lr)

    x = Conv2D(num_filters, kernel_size=9, padding='same')(x)
    x = x_1 = PReLU(shared_axes=[1, 2])(x)

    for _ in range(num_res_blocks):
        x = residual_block(x, num_filters)

    x = Conv2D(num_filters, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Add()([x_1, x])

    for _ in range(num_upsamples):
        x = upsample(x, num_filters * 4)

    x = Conv2D(3, kernel_size=9, padding='same', activation='tanh')(x)
    sr = Lambda(denormalize)(x)


    return Model(lr, sr)
```

```python
def discriminator_block(x_in, num_filters, strides=1, batchnorm=True, momentum=0.8):
    x = Conv2D(num_filters, kernel_size=3, strides=strides, padding='same')(x_in)
    if batchnorm:
        x = BatchNormalization(momentum=momentum)(x)
    return LeakyReLU(alpha=0.2)(x)


def build_discriminator(hr_shape):
    x_in = Input(shape=(hr_shape, hr_shape, 3))
    x = Lambda(normalize)(x_in)

    x = discriminator_block(x, 64, batchnorm=False)
    x = discriminator_block(x, 64, strides=2)

    x = discriminator_block(x, 128)
    x = discriminator_block(x, 128, strides=2)
    
    x = Flatten()(x)

    x = Dense(32)(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dense(1, activation='sigmoid')(x)

    return Model(x_in, x)
```

```python
generator = build_srresnet()
generator.summary()
```

```python
discriminator = build_discriminator(128)
discriminator.summary()
```

```python
bce = BinaryCrossentropy()
mse = MeanSquaredError()
learning_rate=PiecewiseConstantDecay(boundaries=[100000], values=[1e-4, 1e-5])

gen_optimizer = Adam(learning_rate=learning_rate)
disc_optimizer = Adam(learning_rate=learning_rate)
```

```python
def psnr_metric(x1, x2):
    return tf.image.psnr(x1, x2, max_val=255)

checkpoint_path = '../scratch/lung_SRGAN/checkpoints/'

srgan_checkpoint = tf.train.Checkpoint(step=tf.Variable(0),
                                       psnr=tf.Variable(0.0),
                                       generator_optimizer=Adam(learning_rate),
                                       discriminator_optimizer=Adam(learning_rate),
                                       generator=generator,
                                       discriminator=discriminator)

srgan_checkpoint_manager = tf.train.CheckpointManager(srgan_checkpoint, checkpoint_path, max_to_keep=5)
```

```python
@tf.function
def train_step(lr, hr):
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        lr = tf.cast(lr, tf.float32)
        hr = tf.cast(hr, tf.float32)
        
        sr = srgan_checkpoint.generator(lr, training = True)
        
        hr_output = srgan_checkpoint.discriminator(hr, training = True)
        sr_output = srgan_checkpoint.discriminator(sr, training = True)
        
        con_loss = calculate_content_loss(hr, sr)
        gen_loss = calculate_generator_loss(sr_output)
        perc_loss = con_loss + 0.001 * gen_loss
        disc_loss = calculate_discriminator_loss(hr_output, sr_output)
    
    gen_gradients = gen_tape.gradient(perc_loss, srgan_checkpoint.generator.trainable_variables)
    disc_gradients = disc_tape.gradient(disc_loss, srgan_checkpoint.discriminator.trainable_variables)
    
    gen_optimizer.apply_gradients(zip(gen_gradients, srgan_checkpoint.generator.trainable_variables))
    disc_optimizer.apply_gradients(zip(disc_gradients, srgan_checkpoint.discriminator.trainable_variables))
    
    return perc_loss, disc_loss
    
@tf.function
def calculate_content_loss(hr, sr):
    sr = preprocess_input(sr)
    hr = preprocess_input(hr)
    sr_features = perceptual_model(sr) / 12.75
    hr_features = perceptual_model(hr) / 12.75
    return mse(hr_features, sr_features)

def calculate_generator_loss(sr_out):
    return bce(tf.ones_like(sr_out), sr_out)

def calculate_discriminator_loss(hr_out, sr_out):
    hr_loss = bce(tf.ones_like(hr_out), hr_out)
    sr_loss = bce(tf.zeros_like(sr_out), sr_out)
    return hr_loss + sr_loss
```

```python
perceptual_loss_metric = Mean()
discriminator_loss_metric = Mean()

num_epochs = 100  # Specify the number of epochs
steps = len(train_hr_generator)
early_stopping_patience = 15  # Specify the patience for early stopping


# Set up TensorBoard logging
log_dir = "../scratch/lung_SRGAN/tensorboard/"  # Modify the log directory as needed
tensorboard_callback = TensorBoard(log_dir=log_dir)

monitor_folder = "../scratch/lung_SRGAN/monitor"

srgan_checkpoint.step.assign(0)

for epoch in range(num_epochs):
    
    epoch_step_counter = 1
    
#     print(f"Epoch {epoch+1} starting")
            
    for lr_batch, hr_batch in zip(train_lr_generator, train_hr_generator):
        srgan_checkpoint.step.assign_add(1)
        epoch_step_counter += 1
        step = srgan_checkpoint.step.numpy()
        
#         print(f"Step {epoch_step_counter}")
        
        now = time.perf_counter()
        
        perceptual_loss, discriminator_loss = train_step(lr_batch, hr_batch)
        perceptual_loss_metric(perceptual_loss)
        discriminator_loss_metric(discriminator_loss)
        
        psnr = 0.0
        
        if step % 100 == 0:
            psnr_values = []

            for lr_batch, hr_batch in zip(val_lr_batches, val_hr_batches):
                sr_batch = srgan_checkpoint.generator.predict(lr_batch)
                sr_batch = tf.clip_by_value(sr_batch, 0, 255)
                sr_batch = tf.round(sr_batch)
                sr_batch = tf.cast(sr_batch, tf.uint8)

                for sr, hr in zip(sr_batch, hr_batch):
                    psnr_value = psnr_metric(hr, sr).numpy()
                    psnr_values.append(psnr_value)
                psnr = tf.reduce_mean(psnr_values)

            image = Image.fromarray(sr_batch.numpy()[0])
            image.save(f"{monitor_folder}/epoch_{epoch}-step_{epoch_step_counter}.png")
            
        
            duration = time.perf_counter() - now

            print(f'Epoch {epoch+1}/{num_epochs}, Step {epoch_step_counter}/{steps}, psnr = {psnr}, perceptual loss = {perceptual_loss_metric.result():.4f}, discriminator loss = {discriminator_loss_metric.result():.4f} ({duration:.2f}s)')

            perceptual_loss_metric.reset_states()
            discriminator_loss_metric.reset_states()

            srgan_checkpoint.psnr.assign(psnr)
            srgan_checkpoint_manager.save()
            
            latest_checkpoint = srgan_checkpoint_manager.latest_checkpoint
            print(f'Latest Checkpoint: {latest_checkpoint}')
            
        if epoch_step_counter >= steps:
            break
            

    # Log to TensorBoard at the end of each epoch
    with tf.summary.create_file_writer(log_dir).as_default():
        tf.summary.scalar('perceptual_loss', perceptual_loss_metric.result(), step=epoch)
        tf.summary.scalar('discriminator_loss', discriminator_loss_metric.result(), step=epoch)
        tf.summary.scalar('psnr', psnr, step=epoch)

# Make sure lose the TensorBoard writer at the end
tensorboard_callback.on_epoch_end(epoch, logs=None)
        
```
